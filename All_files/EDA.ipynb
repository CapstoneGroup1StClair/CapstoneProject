```python
import pandas as pd

# Load the Excel file into a pandas dataframe
df = pd.read_csv('projects.csv', low_memory = False)

# Display the first few rows of the dataframe
df.head()
```

```python
l = []
for year in range(1996, 2023):
    l.append(str(year) + ".3")
l
```

```python
data_issue = df[['Scope','1996.3',
 '1997.3',
 '1998.3',
 '1999.3',
 '2000.3',
 '2001.3',
 '2002.3',
 '2003.3',
 '2004.3',
 '2005.3',
 '2006.3',
 '2007.3',
 '2008.3',
 '2009.3',
 '2010.3',
 '2011.3',
 '2012.3',
 '2013.3',
 '2014.3',
 '2015.3',
 '2016.3',
 '2017.3',
 '2018.3',
 '2019.3',
 '2020.3',
 '2021.3',
 '2022.3']]
data_retire = df[['Scope', '1996.2',
 '1997.2',
 '1998.2',
 '1999.2',
 '2000.2',
 '2001.2',
 '2002.2',
 '2003.2',
 '2004.2',
 '2005.2',
 '2006.2',
 '2007.2',
 '2008.2',
 '2009.2',
 '2010.2',
 '2011.2',
 '2012.2',
 '2013.2',
 '2014.2',
 '2015.2',
 '2016.2',
 '2017.2',
 '2018.2',
 '2019.2',
 '2020.2',
 '2021.2',
 '2022.2']]
```

```python
data_issue = data_issue.melt(id_vars=['Scope'], var_name='year_issue', value_name='Credits_issued')
data_issue
```

```python
data_retire = data_retire.melt(id_vars=['Scope'], var_name='year_retire', value_name='Credits_retired')
data_retire
```

```python
data_issue['year_issue'] = data_issue['year_issue'].str.split('.').str[0].astype(int)
data_issue
```

```python
data_retire['year_retire'] = data_retire['year_retire'].str.split('.').str[0].astype(int)
data_retire
```

```python
data_clean = pd.concat([data_issue, data_retire], axis = 1)
data_clean
```

```python
data_clean.to_csv('clean_data1.csv')
```
