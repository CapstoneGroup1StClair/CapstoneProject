```python
train_size = int(len(weekly) * 0.8)
val_size = int(len(weekly) * 0.1)
train_data = weekly.iloc[:train_size]
val_data = weekly.iloc[train_size: train_size + val_size]
test_data = weekly.iloc[train_size + val_size:]
```

```python
import numpy as np
#train_data = scaler.fit_transform(train_data)
#val_data = scaler.transform(val_data)
#test_data = scaler.transform(test_data)
train_data = np.array(train_data)
val_data = np.array(val_data)
test_data = np.array(test_data)
```

```python
import numpy as np
def generate_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        try:
            X.append(data[i:i+seq_length])
            y.append(data[i+seq_length])
        except KeyError:
            break
    return np.array(X), np.array(y)
```

```python
seq_length = 2
X_train, y_train = generate_sequences(train_data, seq_length)
X_val, y_val = generate_sequences(val_data, seq_length)
X_test, y_test = generate_sequences(test_data, seq_length)
```

```python
X_test, y_test = generate_sequences(test_data, seq_length)
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, LSTM, BatchNormalization, Bidirectional, Flatten
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=64, batch_input_shape=(1,2,1),  stateful = True, return_sequences = True))
model.add(BatchNormalization())
model.add(Dropout(0.25))
model.add(LSTM(units=32,  stateful = True))
model.add(BatchNormalization())
model.add(Dropout(0.25))
model.add(Dense(units=1 ))
model.summary()
```

```python
import tensorflow
model.compile(optimizer=tensorflow.keras.optimizers.Adam(lr = 0.005), loss='mean_squared_error', metrics = ['RootMeanSquaredError'])
stop = EarlyStopping(monitor='val_loss', mode = 'min', patience = 50)
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', verbose=1, save_best_only=True)
```

```python
model.fit(X_train, y_train, epochs=300, verbose=1, batch_size = 1, shuffle = False, validation_data = (X_val, y_val), callbacks = [stop, mc])
```

```python
best_model = load_model('best_model.h5')
```

```python
y_pred_arr = []
for i in range(0, 21, 1):
    x_test = X_test[i:i+1]
    y_pred = best_model.predict(x_test)
    y_pred_arr.append(y_pred)
y_pred_arr = np.array(y_pred_arr)
y_pred_arr = y_pred_arr.reshape(21, -1)
y_pred_arr.shape
```
